```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import scipy
```

```python
def getColstatistic(df, tag):
    # Calculate mean and standard deviation for each row
    means = df.mean(axis=1)
    std_devs = df.std(axis=1)

    # Add mean and standard deviation columns to the DataFrame
    df[f'{tag}_mean'] = means/10000
    df[f'{tag}_std'] = std_devs/10000
    return df.iloc[:,-2:]


def getColstatisticBySearch(df, search_tag, out_tag):
    # print(df.shape)
    column_names = list(df.columns)
    col_find = []
    for col in column_names:
        if re.search(search_tag, col):
            # print(f'{col} find search_tag {search_tag}')
            col_find.append(col)
    
    df_out = df[col_find]
    print(f"df out shape is {df_out.shape}")
    means = df_out.mean(axis=1)
    std_devs = df_out.std(axis=1)
    skew_v = df_out.skew(axis=1)
    kurt_v = df_out.kurt(axis=1)
    df_out[f'{out_tag}_mean'] = means/10000
    df_out[f'{out_tag}_std'] = std_devs/10000
    df_out[f'{out_tag}_skew'] = skew_v/10000
    df_out[f'{out_tag}_kurt'] = kurt_v/10000
    print(df_out.shape)
    
    return df_out.iloc[:,-4:]
```

```python
df = pd.read_csv("../Vegetation_Indices/NDVI_EVI.csv")
df_evi2005 = df.loc[:,'evi_2005-01':'evi_2005-12']
df_evi2005 = getColstatistic(df=df_evi2005, tag='evi2005')

df_evi2006 = df.loc[:,'evi_2006-01':'evi_2006-12']
df_evi2006 = getColstatistic(df=df_evi2006, tag='evi2006')

df_evi2007 = df.loc[:,'evi_2007-01':'evi_2007-12']
df_evi2007 = getColstatistic(df=df_evi2007, tag='evi2007')

df_evi2008 = df.loc[:,'evi_2008-01':'evi_2008-12']
df_evi2008 = getColstatistic(df=df_evi2008, tag='evi2008')

df_evi_jan = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-01", out_tag="evi_jan")
df_evi_feb = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-02", out_tag="evi_feb")
df_evi_mar = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-03", out_tag="evi_mar")
df_evi_apr = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-04", out_tag="evi_apr")
df_evi_may = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-05", out_tag="evi_may")
df_evi_jun = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-06", out_tag="evi_jun")
df_evi_jul = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-07", out_tag="evi_jul")
df_evi_aug = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-08", out_tag="evi_aug")
df_evi_sep = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-09", out_tag="evi_sep")
df_evi_oct = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-10", out_tag="evi_oct")
df_evi_nov = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-11", out_tag="evi_nov")
df_evi_dec = getColstatisticBySearch(df=df, search_tag=r"evi_\d\d\d\d-12", out_tag="evi_dec")

evi_list = [df_evi2005, df_evi2006, df_evi2007, df_evi2008, 
            df_evi_jan, df_evi_feb, df_evi_mar, df_evi_apr, df_evi_may, df_evi_jun, 
            df_evi_aug, df_evi_sep, df_evi_oct, df_evi_nov, df_evi_dec]
evi_stat = pd.concat(evi_list, axis=1)
evi_stat

```
